# Уравнение пьезопроводности для плоскорадиальной симметрии имеет вид

$$
\frac{\partial^2p}{\partial t^2} + \frac{1}{r}\frac{\partial p}{\partial r} = \frac{1}{\kappa}\frac{\partial p}{\partial t}
$$
где 
- $p(r,t)$ - давление в пласте
- $r$ - расстояние
- $t$ - время
- $\kappa$ -  коэффициент пьезопроводности 
Для построения решения необходимо задать начальное и граничные условия. 

Начальное условие - равенство давления начальному $p_i$ в момент времени $t=0$
$$
\left. p(r,t) \right|_{t=0} = p_i
$$
Граничное условие на скважине

---

Дебит на забое скважины, можно определить из закона Дарси
$$
\left.q(r,t) \right|_{r=r_w} = - \left. \frac{2\pi kh}{\mu B} \left( r \frac{\partial p}{\partial r} \right) \right|_{r=r_w}
$$

---
Введем безразмерные переменные
$$
p_D = \frac{2\pi kh (p_i - p)}{ q_{ref}\mu B }
$$
$$
t_D=\frac{\kappa t}{r_w^2}
$$
$$
q_D=\frac{q}{q_{ref}}
$$
$$
r_D = \frac{r}{r_w}
$$

---

В безразмерных переменных выражение для дебита на забое скважины примет вид
$$
q_D = r_D \frac{\partial p_D}{\partial r_D}
$$

---

Тогда уравнение пьезопроводности примет вид

$$
\frac{\partial^2p_D}{\partial r_D^2} + \frac{1}{r_D}\frac{\partial p_D}{\partial r_D} = \frac{\partial p_D}{\partial t_D}  , r_D>0, t_D,>0
$$

начальное условие

$$
p_D=0, t_D=0
$$

Граничные условия на скважине для задачи с постоянным дебитом

$$
2\pi r_D \frac{\partial p_D}{\partial r_D} = С_{qD}
$$

Еще есть какое то решение на бесконечности или внешнем контуре. Например для бесконечности оно будет иметь вид:

$$
\left. p_D \right|_{r_D -> \infty} = 0
$$

---

Эта задача записывается в пространстве Лапласа в следующем виде

$$
\frac{d^2 \widetilde{p_D} }{dr_D^2} = \frac{1}{r_D} \frac{d \widetilde{p_D} }{dr_D} = u \widetilde{p_D}
$$
где
- $\widetilde{p_D}(u,r_D)$ - образ давления  в пространстве Лапласа
- $u$ - переменная в пространстве Лапласа
- $r_D$ - безразмерное расстояние
граничные условия для постоянного дебита примут вид
$$
r_D \frac{d\widetilde{p_D}}{dr_D} = \frac{С_{qD}}{u}
$$
$$
\left. \widetilde{p_D} \right |_{r_D -> \infty} = 0
$$
Решение будет иметь вид
$$
\widetilde{p_D} = A K_0 + BI_0
$$
Пусть решение такой задачи известно и выражается как $\widetilde{p_D}_{q.const}$ 

---




Запишем задачу для постоянного забойного давления. Для нее граничные условия в пространстве Лапласа примут вид

$$
\left. \widetilde{p_D} \right|_{r_D=1} = \frac{C_{pD}}{u} \tag{1}
$$

Для такой задачи решение будем искать в виде

$$
\widetilde{p_D}(u, r_D) = \widetilde{p_D}_{p.const}(u, p_D) =  \alpha(u)\widetilde{p_D}_{q.const}(u, r_D)
$$

Для такого решения условия на бесконечности или на внешней границе будут также удовлетворяться (условия постоянства давления или непротекания).

$$
\left.  \alpha(u)\widetilde{p_D}_{q.const}(u)\right|_{r_D=1} = \frac{C_{pD}}{u}
$$

$$
\alpha(u)= \frac{C_{pD}}{u \left.\widetilde{p_D}_{q.const} \right|_{r_D=1}}
$$

Подставим этот вид решения в условие на границе (1)

$$
\widetilde{p_D}(u, r_D) = \widetilde{p_D}_{p.const}(u, r_D) =\alpha(u) \widetilde{p_D}_{q.const}(u, r_D) = \frac{C_{pD}}{u} \frac{\widetilde{p_D}_{q.const}(u, r_D)}{\left. \widetilde{{p_D}}_{q.const}(u)\right|_{r_D=1}}
$$




Соответственно выражение для безразмерного дебита в пространстве Лапласа примет вид


$$
q_D = r_D \frac{\partial p_D}{\partial r_D}
$$

$$
\widetilde{q_D} = r_D    \frac{C_{pD}}{u\left. \widetilde{{p_D}}_{q.const}(u)\right|_{r_D=1}}  \frac{\partial \widetilde{p_D}_{q.const}(u, r_D) }{\partial r_D}
$$

$$
\widetilde{q_D} =     \frac{C_{pD}}{u\left. \widetilde{{p_D}}_{q.const}(u)\right|_{r_D=1}}  \frac{C_{qD}}{ u} 
$$

$$
\widetilde{q_D} = -\frac{C_{pD }C_{qD}}{ u^2}   \frac{1}{\left. \widetilde{{p_D}}_{q.const}(u)\right|_{r_D=1}}  
$$
