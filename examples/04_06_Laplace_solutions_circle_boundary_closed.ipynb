# 1 Уравнение фильтрации <a class="anchor" id="461"></a>

Рассматривается уравнение фильтрации в безразмерных переменных 

$$ 
\frac{\partial p_D}{ \partial t_D} = \frac{1}{r_D}\left[ \frac{ \partial{}}{ \partial{r_D} }\left( r_D \dfrac{\partial p_D}{ \partial r_D} \right) \right]  
\tag{1.1}
$$

где введены следующие безразмерные переменные
* $r_D$ - безразмерное расстояние от центра скважины
* $t_D$ - безразмерное время
* $p_D$ - безразмерное давление

Соответствующие определения безразмерных переменных

$$ r_D = \frac{r}{r_w} \tag{1.2}$$
$$ t_D = \frac{0.00036 kt}{\phi \mu c_t r_w^2}  \tag{1.3}$$
$$ p_D = \frac{kh}{ 18.41 q_s B \mu} \left( p_i - p \right)  \tag{1.4} $$

где в свою очередь 
* $q_s$ - дебит скважины на поверхности, приведенный к нормальным условиям, м3/сут
* $\phi$ - пористость, доли единиц
* $\mu$ - вязкость нефти в пласте, сП
* $B$ - объемный коэффициент нефти, м3/м3
* $p_i$ - начальное давление в пласте, атм
* $p$ - давление забойное, атм
* $c_t$ - общая сжимаемость системы в пласте, 1/атм
* $k$ - проницаемость, мД
* $t$ -  время, час
* $r$ - расстояние от центра скважины, м
* $r_w$ - радиус скважины, м

Решение такого уравнение может быть получено с использованием <a href="https://ru.wikipedia.org/wiki/%D0%9F%D1%80%D0%B5%D0%BE%D0%B1%D1%80%D0%B0%D0%B7%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5_%D0%9B%D0%B0%D0%BF%D0%BB%D0%B0%D1%81%D0%B0"
                        target="_blank" rel="noreferrer noopener">преобразования Лапласа</a>.

$$ 
L \left [ f(t) \right] = \tilde{f}(u) = \int_{0}^{\infty}f(t)e^{-ut}dt 
\tag{1.4}
$$

где $u$ параметр пространства Лапласа соответствующий времени.

Тогда уравнение в пространстве Лапласа преобразуется к виду:

$$ 
u \widetilde{p_D}  =  \dfrac{1}{r_D} \left[\dfrac{d}{d r_D} \left(r_D \dfrac{d{\widetilde{p_D}}}{d r_D} \right) \right] 
\tag{1.5}
$$

Уравнение известно как  <a href="https://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%B4%D0%B8%D1%84%D0%B8%D1%86%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%BD%D1%8B%D0%B5_%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D0%B8_%D0%91%D0%B5%D1%81%D1%81%D0%B5%D0%BB%D1%8F"> модифицированное уравнение Бесселя </a>. 
где 
* $u$ - переменная пространства Лапласа, соответствующая времени
* $\widetilde{p_D}(u, r_D)$ - изображение давления в пространстве Лапласа

С уравнением  можно начать работать с использованием системы компьютерной алгебры [`sympy`](https://www.sympy.org/en/index.html). 

# 2 Вывод решения с круговой границей постоянного давления<a class="anchor" id="462"></a>

```python
# определим переменные с которыми будем работать 

t_d, r_d, u = sp.symbols('t_D r_D u', real=True, positive=True)
p_d, Lp_d = sp.symbols('p_D Lp_D', cls=sp.Function, real=True, positive=True)
r_ed = sp.symbols('r_eD', positive=True)

# зададим в явном виде решение с использованием K_0 и I_0
A, B = sp.symbols('A B')
eq_sol_general = sp.Eq(Lp_d(r_d) , A * sp.besselk(0, r_d * sp.sqrt(u)) + B * sp.besseli(0, r_d * sp.sqrt(u)))
display(eq_sol_general)
```

```python
# определим уравнение фильтрации в пространстве Лапласа
diff_eq = sp.Eq(u * p_d(r_d),
                1 / r_d * (sp.diff(r_d * sp.diff(p_d(r_d), r_d) , r_d)) )
diff_eq
```

## 6 Частное решение для конечного круглого пласта с постоянным давлением на контуре и конечного радиуса скважины. <a class="anchor" id="426"></a>

Частное решение строится за счет поиска параметров $A$ и $B$ для уравнения (2.5) так, чтобы они удовлетворяли граничным условиям. 

1. Граничное условие на закрытом внешнем контуре  

$$\lim_{r_D \to r_{eD}} \left[ \dfrac{ \partial p_D(r_D, t_D)}{\partial r_D} \right] = 0 $$

в пространстве Лапласа преобразуется в следующее

$$\lim_{r_D \to r_{eD}} \left[ \dfrac{ d \widetilde{p_D}}{d r_D} \right] = 0 $$ 

2. Граничное условие на скважине

$$\lim_{r_D \to r_{wD}} \left[ r_D \dfrac{ \partial p_D(r_D, t_D)}{\partial r_D} \right] = -1 $$

в пространстве Лапласа  преобразуется к

$$\lim_{r_D \to r_{wD}} \left[ r_D \dfrac{ d \widetilde{p_D}}{d r_D} \right] = -\dfrac{1}{u} $$ 


```python
# граничное условие на скважине
eq_boundary = sp.Eq(r_d* sp.diff(p_d(r_d,t_d),r_d) , -1)
display(eq_boundary)
```

```python
# применим преобразование Лапласа к обеим частям граничного условия
eq_boundary_Laplace = sp.Eq(sp.laplace_transform(eq_boundary.lhs, t_d, u,  noconds=True) ,  
                            sp.laplace_transform(eq_boundary.rhs, t_d, u,  noconds=True))
display(eq_boundary_Laplace)
```

```python
# выведем общее решение уравнения фильтрации в радиальной форме
eq_sol_general
```

```python
# запишем первое граничное условие на границе 
# постоянное давление или нулевой перепад давления
eq1_boundary_Laplace = sp.Eq(eq_sol_general.rhs.diff(r_d).subs(r_d, r_ed),0)
display(eq1_boundary_Laplace)
```

```python
# запишем второе граничное условие на скважине
# условие постоянства дебита
eq2_boundary_Laplace = sp.Eq(r_d * eq_sol_general.rhs.diff(r_d), -1/u).subs(r_d, 1)
eq2_boundary_Laplace
```

```python
# решим уравнения граничных условий относительно A, B
sol_AB = sp.solve([eq1_boundary_Laplace,eq2_boundary_Laplace], [A,B])
display(sol_AB)
display(sp.Eq(A, sol_AB[A]))
display(sp.Eq(B, sol_AB[B]))
```

```python
sol_re_closed = eq_sol_general.subs(A, sol_AB[A]).subs(B, sol_AB[B]).simplify()  
sol_re_closed
```

Таким образом получено решение для замкнутой круговой границы вокруг скважины 

$$
\displaystyle \operatorname{Lp}_{D}{\left(r_{D} \right)} = \frac{I_{0}\left(r_{D} \sqrt{u}\right) K_{1}\left(r_{eD} \sqrt{u}\right) + I_{1}\left(r_{eD} \sqrt{u}\right) K_{0}\left(r_{D} \sqrt{u}\right)}{u^{\frac{3}{2}} \left(- I_{1}\left(\sqrt{u}\right) K_{1}\left(r_{eD} \sqrt{u}\right) + I_{1}\left(r_{eD} \sqrt{u}\right) K_{1}\left(\sqrt{u}\right)\right)}$$

Ипользуя подход показанный выше можно построить графики зависимости безразмерного давления от времени и расстояния для этого решения с использованием numpy и scipy функций

# 3. Построение графиков используя сгенерированное решение <a class="anchor" id="463"></a>

```python
# преобразование sympy выражения в lambda функцию с использованием numpy
soln_re_closed_numpy = sp.lambdify([u, r_d, r_ed], sol_re_closed.rhs, modules = ['numpy', 'scipy'])

# после преобразования lambda функцию можно использовать для проведения расчетов как обычныую функцию

# определим функцию для расчета обратного преобразования Лапласа
# отдельная функция удобно чтобы передавать r_d второй параметр
def soln_re_closed_inv_lapl_numpy(t_d=1000, r_d=1, r_ed=100):
    sol_lap = anaflow.get_lap_inv(soln_re_closed_numpy,  r_D=r_d, r_eD=r_ed)
    return sol_lap(t_d)

# зададим диапазоны изменения параметров для построения графиков
td_arr = np.logspace(1,5,100)
rd_arr = np.logspace(1,2,100)
re=100
# построим графики
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[15,5])
for ri in np.linspace(1,100,10):
    ax1.plot(td_arr, soln_re_closed_inv_lapl_numpy(td_arr, r_d=ri, r_ed=re), label = f'$r_d={ri:.0f}$')
for ti in np.linspace(1e2, 1e4,10):
    ax2.plot(rd_arr, np.squeeze( [soln_re_closed_inv_lapl_numpy(ti, r_d=ri, r_ed=re) for ri in rd_arr ] ), label = f'$t_d={ti:.0f}$')
# наведем немного красоты
ax1.set_xlabel('$t_d$')
ax1.set_ylabel('$p_d$')
ax1.legend()
ax2.set_xlabel('$r_d$')
ax2.set_ylabel('$p_d$')
ax2.legend()
plt.show()
```

# 4. Построение графиков используя явно заданное решение <a class="anchor" id="464"></a>

```python
# Зададим в явном виде решение в пространстве Лапласа
def pd_laplace_circle_closed(u, r_d=1, r_ed=1000):
    """
    расчет решения в пространстве Лапласа для замкнутой круговой границы
    u -  переменная пространства Лапласа
    r_ed - безразмерный радиус контура
    """
    sqrtu = u ** 0.5
    r_ed_sqrtu = r_ed * sqrtu
    
    pdl = ((sc.iv(0, r_d*sqrtu)*sc.kv(1, r_ed_sqrtu) + 
           sc.iv(1, r_ed_sqrtu)*sc.kv(0, r_d*sqrtu)) / 
           (u**(3/2)*(-sc.iv(1, sqrtu)*sc.kv(1, r_ed_sqrtu) + 
                      sc.iv(1, r_ed_sqrtu)*sc.kv(1, sqrtu))))
    return pdl

# построим функцию инвертирующую решение в пространстве Лапласа

# функция расчета безразмерного давления с использованием алгоритма Стефеста
# для численного обратного преобразования Лапласа
def pd_inv(pd_lapl, t_d, r_d, **kwargs):
        pd_inv_ = anaflow.get_lap_inv(pd_lapl, r_d=r_d, **kwargs)
        return pd_inv_(t_d)
```

```python
pd_inv(pd_laplace_circle_closed, 100, 1)
```

```python
# зададим диапазоны изменения параметров для построения графиков
td_arr = np.logspace(1,5,100)
rd_arr = np.logspace(1,2,100)
re=100
# построим графики
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[15,5])
for ri in np.linspace(1,100,10):
    ax1.plot(td_arr, pd_inv(pd_laplace_circle_closed, td_arr, r_d=ri, r_ed=re), label = f'$r_d={ri:.0f}$')
for ti in np.linspace(1e2, 1e4,10):
    ax2.plot(rd_arr, np.squeeze( [pd_inv(pd_laplace_circle_closed, ti, r_d=ri, r_ed=re) for ri in rd_arr ] ), label = f'$t_d={ti:.0f}$')
# наведем немного красоты
ax1.set_xlabel('$t_d$')
ax1.set_ylabel('$p_d$')
ax1.legend()
ax2.set_xlabel('$r_d$')
ax2.set_ylabel('$p_d$')
ax2.legend()
plt.show()
```

# 5. Построение решения с использованием `sympy.dsolve` <a class="anchor" id="465"></a>

Отметим, что в некоторых случаях можно построить частные решения для ОДУ напряму с использованием `sympy`

Вспомним с каким уравнением мы работаем

```python
diff_eq
```

Построим снова общее решение, чтобы убедиться, что на текущем этапе все работает

```python
soln = sp.dsolve(diff_eq, p_d(r_d))
soln
```

Построим частное решение сразу задав в `sympy.dsolve` граничные условия на искомую функцию

```python
soln_sympy = sp.dsolve(diff_eq, p_d(r_d), 
                       ics={p_d(r_d).diff(r_d).subs(r_d, r_ed):0, 
                            p_d(r_d).diff(r_d).subs(r_d,1):-1/u}).simplify()
soln_sympy
```

Решение построено, хотя оно выражается через $Y_0$ вместо $K_0$ в решении (6.5) и в нем в явном виде присутствует $i$ мнимая единица. 

Это решение также можно использовать для проведения расчетов и построения графиков. Для того, чтобы подавить предупреждения об игнорировании мнимой части - при проведении обратного преобразования Лапласа оставим только реальную часть в явном виде.

Также ограничим диапазон отображения графика зависимости давления от времени - чтобы не показывать неверные значения получаемых в ходе расчета для малых времен.

```python
# преобразование sympy выражения в lambda функцию с использованием numpy
soln_re_const_p_numpy = sp.lambdify([u, r_d, r_ed], soln_sympy.rhs, modules = ['numpy', 'scipy'])

# после преобразования lambda функцию можно использовать для проведения расчетов как обычныую функцию

# определим функцию для расчета обратного преобразования Лапласа
# отдельная функция удобно чтобы передавать r_d второй параметр
def soln_re_const_p_inv_lapl_numpy(t_d=1000, r_d=1, r_e=100):
    sol_lap = anaflow.get_lap_inv(soln_re_const_p_numpy,  r_D=r_d, r_eD=r_e)
    return sol_lap(t_d).real

# зададим диапазоны изменения параметров для построения графиков
td_arr = np.logspace(2,5,100)
rd_arr = np.logspace(1,2,100)
re=100
# построим графики
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[15,5])
for ri in np.linspace(2,100,10):
    ax1.plot(td_arr, soln_re_const_p_inv_lapl_numpy(td_arr, r_d=ri, r_e=re), label = f'$r_d={ri:.0f}$')
for ti in np.linspace(2e2, 1e4,10):
    ax2.plot(rd_arr, np.squeeze( [soln_re_const_p_inv_lapl_numpy(ti, r_d=ri, r_e=re) for ri in rd_arr ] ), label = f'$t_d={ti:.0f}$')
# наведем немного красоты
ax1.set_xlabel('$t_d$')
ax1.set_ylabel('$p_d$')
ax1.set_ylim(0,15)
ax1.legend()
ax2.set_xlabel('$r_d$')
ax2.set_ylabel('$p_d$')
ax2.legend()
plt.show()
```

Графики построились, хотя и пришлось подавлять предупреждения об игнорировании мнимой части и для одного из значений расстояния расчет развалился. Первый вариант построения решения с использованием $K_0$ выглядит надежнее.
